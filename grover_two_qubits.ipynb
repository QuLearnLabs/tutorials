# Grover's algorithm with 2 qubits

## Initialise a 2 qubit quantum circuit

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler
from qiskit.visualization import plot_histogram

qc = QuantumCircuit(2)

qc.draw("mpl")
```

## Put all qubits into superposition

```python
qc.h(0)
qc.h(1)

qc.draw("mpl")
```

## Apply the oracle

```python
qc.cz(0, 1)

qc.draw("mpl")
```

## Perform the Grover diffusion

```python
qc.h(0)
qc.h(1)

qc.x(0)
qc.x(1)

qc.cz(0, 1)

qc.x(0)
qc.x(1)

qc.h(0)
qc.h(1)

qc.draw("mpl")
```

## Measure the state

```python
qc.measure_all()

qc.draw("mpl")
```

## Execute the circuit with the sampler

```python
shots = 1000

result = StatevectorSampler().run([qc], shots=shots).result()

counts = result[0].data.meas.get_counts()

plot_histogram({p: c / shots for p, c in counts.items()})
```
